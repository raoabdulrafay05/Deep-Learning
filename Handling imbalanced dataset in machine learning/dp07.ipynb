Handling imbalanced dataset in machine learning

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("Churn_Modelling.csv")
df1
```

```python
df1 = df1.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
df1
```

```python
df1['HasCrCard'].value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler1 = MinMaxScaler()

columns_to_scale_1 = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
df1[columns_to_scale_1] = scaler1.fit_transform(df1[columns_to_scale_1])
```

```python
df1 = pd.get_dummies(data = df1, columns = ['Geography'])
```

```python
df1 = df1.replace({True:1, False:0})
df1
```

```python
df1.replace({'Female' : 0, 'Male' : 1}, inplace=True)
```

```python
X1 = df1.drop(columns=['Exited'])
y1 = df1.Exited
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1,y1,test_size=0.2,random_state=42, stratify=y1)
```

```python
model = keras.Sequential([
    keras.layers.Dense(7, input_shape = (12,), activation='relu'),
    keras.layers.Dense(3, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train1, y_train1, epochs=100)
```

```python
model.evaluate(X_test1,y_test1)
```

```python
y_pred = np.round(model.predict(X_test1).reshape(-1))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test1, y_pred))
```

**Now we have to handle imbalanced data**

```python
def ANN(X_train, X_test, y_train, y_test, loss, weights):
    
    model = keras.Sequential([
    keras.layers.Dense(7, input_shape = (12,), activation='relu'),
    keras.layers.Dense(3, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
    ])
    
    model.compile(
    optimizer = 'adam',
    loss = loss,
    metrics = ['accuracy']
    )
    
    if weights == -1:
        model.fit(X_train, y_train, epochs = 100)
    else:
        model.fit(X_train,y_train, class_weight=weights, epochs=100)
    
    print(model.evaluate(X_test,y_test))
    
    y_preds = np.round(model.predict(X_test).reshape(-1))
    
    print("classification_report: \n", classification_report(y_test,y_preds))
    
    return y_preds
    
```

***SMOTE METHOD OF DEALING WITH IMBALANCED DATASET***

```python
X = df1.drop(columns=['Exited'])
y = df1.Exited
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
```

```python
X.shape, y.shape
```

```python
y.value_counts()
```

```python
X_sm, y_sm = smote.fit_resample(X,y)
```

```python
y_sm.value_counts()
```

```python
X_sm.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm,y_sm,test_size=0.2, random_state=42, stratify=y_sm)
```

```python
y_preds = ANN(X_train,X_test,y_train,y_test, 'binary_crossentropy',-1)
```

```python
print("classification_report: \n", classification_report(y_test,y_preds))
```
