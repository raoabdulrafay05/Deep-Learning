Gradient Descent For Neural Network

```python
import numpy as np
import pandas as pd
from tensorflow import keras
import tensorflow as tf
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
df['age'] = df['age']/100
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age','affordibility']], df.bought_insurance, test_size=0.25, random_state=32)
```

```python
model = keras.Sequential([
    keras.layers.Dense(1,input_shape = (2,), activation='sigmoid', kernel_initializer='ones', bias_initializer='zeros')
])

model.compile(
    optimizer = 'SGD',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train, y_train, epochs = 300)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def prediction_function(age,affordibility):
    wighted_sum =  coef[0]*age+ coef[1]*affordibility + intercept
    return sigmoid(wighted_sum)
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted = np.clip(y_predicted, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_predicted) + (1 - y_true) * np.log(1 - y_predicted))
```

```python
def sigmoid_numpy(x):
    return 1/(1+np.exp(-x))
```

```python
def gradient_descent(age, affordibility, y_true, epochs):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)
    
    age = np.array(age)
    affordibility = np.array(affordibility)
    y_true = np.array(y_true)

    
    for i in range(epochs):
        wighted_sum = w1*age +w2*affordibility + bias
        y_predicted = sigmoid((wighted_sum))
        loss = log_loss(y_true,y_predicted)
        
        w1d = (1/n)*np.dot(np.transpose(age),(y_predicted-y_true))   
        w2d = (1/n)*np.dot(np.transpose(affordibility),(y_predicted-y_true))   
        
        bias_d = np.mean(y_predicted-y_true)
        
        w1 = w1 - rate*w1d
        w2 = w2 - rate*w2d
        bias = bias - rate*bias_d
        
        print(f'Epoch{i}, w1: {w1}, w2 : {w2}, bias : {bias}, loss: {loss}')
        
        if loss <= 0.5572:
            break
        
    return w1,w2,bias
```

```python
class myNN:
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.bias = 0
    
    def fit(self, X, y, epochs):
        self.gradient_descent(X['age'], X['affordibility'], y,epochs)
    
    def predict(self, X_test):
        weighted_sum = self.w1*X_test['age'] + self.w2*X_test['affordibility'] + self.bias
        return sigmoid(weighted_sum)
    
    def gradient_descent(self, age, affordibility, y_true, epochs):
        w1 = w2 = 1
        bias = 0
        rate = 0.5
        n = len(age)
        
        age = np.array(age)
        affordibility = np.array(affordibility)
        y_true = np.array(y_true)

        
        for i in range(epochs):
            wighted_sum = w1*age +w2*affordibility + bias
            y_predicted = sigmoid((wighted_sum))
            loss = log_loss(y_true,y_predicted)
            
            w1d = (1/n)*np.dot(np.transpose(age),(y_predicted-y_true))   
            w2d = (1/n)*np.dot(np.transpose(affordibility),(y_predicted-y_true))   
            
            bias_d = np.mean(y_predicted-y_true)
            
            w1 = w1 - rate*w1d
            w2 = w2 - rate*w2d
            bias = bias - rate*bias_d
            
            if i%50 == 0:
                print(f'Epoch{i}, w1: {w1}, w2 : {w2}, bias : {bias}, loss: {loss}')
            
            if loss <= 0.5572:
                print(f'Epoch{i}, w1: {w1}, w2 : {w2}, bias : {bias}, loss: {loss}')
                break
        
        return w1,w2,bias
    
```
